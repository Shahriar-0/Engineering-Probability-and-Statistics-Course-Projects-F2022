# Central Limit Theorem

Have you ever wondered why we study the normal distribution?

Certainly, one reason is that some data are naturally modeled as normal. But why does the normal distribution play such a "central" role in statistical studies? In other words, why is it that even if most people don't know much about probability, they've at least seen a bell curve once? Could it be that simply some data are naturally normal, which brings such fame to the normal distribution?

The fame of the normal distribution comes from the fact that if we sum a large number of independent random variables, the resulting distribution will definitely be normal. The accuracy of this statement comes from the word "large." How many random variables constitute "large"? Mathematically speaking, the precise statement is that the sum of independent random variables tends toward a normal distribution as the number approaches infinity. However, in reality, if the distribution of the variables is not too skewed (unlike income distribution, where very few people have extremely high incomes), the sum of a small number of random variables will result in a distribution very close to normal. This is called the "Central" Limit Theorem.

This is why the normal distribution plays such a central role in statistical studies. For example, we can imagine that students' scores in a probability class are a random variable made up of a sum of many factors: students' interest in math, their academic background, the different resources they choose to study, and so on. Each of these variables may not be normally distributed (in fact, many of them cannot be normal because they are discrete), but the sum of them, which results in the final score, will have a normal distribution.

We know that the binomial random variable results from the sum of a number of independent Bernoulli random variables. According to the above content, as the number of Bernoulli random variables or the number of trials in the binomial random variable increases, the resulting distribution tends toward normal. In this exercise, you will demonstrate this with a chart.

First, generate 10,000 realizations of a binomial distribution with a probability of 0.5 and 1,000,000 trials. Then, plot its histogram.

Next, overlay a normal distribution with the same mean and variance as the binomial distribution on the same chart.

Do you observe any similarities?


```r
size <- 100000
probability <- 0.5

realizations <- rbinom(100000, size, probability)
hist(realizations, freq = F)

par(new=TRUE)

t <- seq(0, size,by = size ** (1/4))
x <- dnorm(t, mean = size * probability , sd = size * probability * (1 - probability))
plot(x)
```
