<div dir='rtl'>
در مطالب تدریسی، آموختید هدف از نمونه‌برداری چیست. به دنباله متغیرهای تصادفی 
$i.i.d$
$(X_1, X_2, \cdots, X_n)$
که از یک جامعه($population$)
آماری با توزیع $F$
انتخاب شده باشند، یک نمونه
($sample$)
از توزیع 
($F$)
می‌گویند.
در این تمرین، به دنبال نشان دادن بی‌غرض
($unbiased$)
بودن میانگین و واریانس نمونه هستیم.
</div>

<div dir='rtl'>
در کد زیر نمرات درس برنامه‌سازی پیشرفته(همان نمراتی که در تمرین قبلی نیز استفاده شده بود) به شما داده شده است.
می‌خواهیم با نمونه‌برداری‌های مختلف از این جامعه و محاسبه میانگین و واریانس نمونه، بی‌غرض بودن این تخمینگرها را بررسی کنیم.
</div>

```r
final_grade <- c(
       19.3, 18.2, 14.1, 15.1, 17.1, 0.0,
       16.2, 17.4, 16.1, 19.2, 14.6, 13.4,
       18.0, 15.7, 14.8, 17.0, 19.6, 18.2,
       0.0, 17.2, 0.0, 18.8, 16.7, 11.0,
       17.5, 19.3, 17.2, 13.5, 0.0, 18.5,
       16.8, 16.4, 12.8, 17.9, 19.4, 12.9,
       12.4, 19.1, 12.3, 18.0, 18.4, 18.5,
       15.2, 0.0, 12.9, 14.1, 17.5, 17.2,
       15.6, 14.9, 16.9, 12.0, 15.9, 18.4,
       17.8, 19.3, 12.0, 15.6, 13.7, 10.0,
       17.6, 19.1, 19.9, 18.2, 18.3, 17.7,
       19.0, 15.5, 15.1, 18.4
)
```

```r
# The below function calculates mean and variance of a sample with size n.
sample_mean_and_variance <- function(n) {
  sampled_grade <- sample(final_grade, n)
  return (
    list(
      mean(sampled_grade), var(sampled_grade)
    )
  )
}
```

```r
# Now, try sampling multiple times and save mean and variance for them
total_samples <- 1000
N <- 20 # Sample size
means <- c()
variances <- c()

for(i in 1:total_samples) {
  result <- sample_mean_and_variance(N)
  means <- c(means, as.numeric(result[1]))
  variances <- c(variances, as.numeric(result[2]))
}
```

<div dir='rtl'>
از آنجایی که تابع $var$
در زبان $R$، واریانس نمونه را محاسبه می‌کند،
برای مشاهده بی‌غرض بودن تخمینگرها، لازم است تابعی برای محاسبه واریانس جامعه بنویسیم. می‌دانیم واریانس نمونه به صورت زیر محاسبه می‌شود
($n$ برابر با اندازه نمونه است):
</div>

$$S^2 = Σ_{i=1}^{n} \frac{(x_i - x̄)^2}{n - 1}$$

```r
# The below function calculates population variance of x.
population_variance <- function(x) {
  n <- length(x)
  return (
    var(x) * (n - 1) / n
  )
}
```

```r
# Check whether these estimators are unbiased or not
cat(mean(means), mean(variances))
```

```r
# Above result should be approximately same as below
cat(mean(final_grade), population_variance(final_grade))
```
