# Arithmetic and Assignment

In the R programming language, `Inf` and `-Inf` stand for positive and negative infinity, respectively. They are the result of storing either a large number or a product that is a result of division by zero.

`NaN`, on the other hand, stands for "Not a Number" and is a logical vector of length 1 that applies to numerical values, as well as real and imaginary parts of complex values, but not to values of integer vector¹. It is used to represent the result of undefined or indeterminate mathematical operations. For example, `0/0` returns `NaN`, as does `Inf - Inf`.


```r
2 + 2
10 - 3
9 / 3
log(2.7182818 ** 2)     # ln (e ^ 2)
5 %% 2                  # modulus operator
10 ^ 2                  # ^ is the exponent operator
120 %/% 11              # integer division
sqrt(81)
```

```r
# VARIABLE <- VALUE
var <- 10
print(var)
# VALUE -> VARIABLE
var ^ 2 -> var
print(var)
var = 20
print(var)
```

* Note that the `<-` operator is the preferred way of assigning symbols to values except when **passing parameters** to functions where only the use of the `=` operator is syntactically correct. Symbol names in R are composed of **alphanumeric characters, underscores, and periods**. Symbol names must begin with a **letter or a period**, in the case of the latter, the period cannot be followed by a digit. Contrast this with C in which identifiers cannot contain periods and can begin with underscores. Also keep in mind that unlike C and many other languages you might be familiar with in which periods are operators used to access the elements of structs or objects, periods have no semantic significance in R.
* [Google’s R Style Guide](https://web.stanford.edu/class/cs109l/unrestricted/resources/google-style.html) suggests the usage of `<-` rather than `=` even though the equal sign is also allowed in R to do exactly the same thing when we assign a value to a variable.

# Vectors

* Vectors are the most basic data structures in R, and they are ubiquitous indeed. In fact, even the single values that we've been working with thus far were actually vectors of length 1. That's why the interactive R console has been printing [1] along with all of our output. R language is 1-indexed, meaning that the first element of a vector is at index 1, not 0 as in many other programming languages.

```r
# The canonical way of building vectors manually is 
# by using the c() function (which stands for combine).
vect <- c(1, 2, 3, 4, 5, 6)
print(vect)
```

## Coercion (Casting)

* Note that if we tried to put character data types into this vector as follows:

```r
vect2 <- c(1, 2, 3, "-", 4, 5, 6)
print(vect2)
```

* R would convert all the items in the vector (called elements) into character data
types to satisfy the condition that all elements of a vector must be of the same
type.

* Another way to create vectors is by using **sequences**.

```r
vect <- 1:10
print(vect)
```

* Another way is to use `seq()` function.

```r
# seq(x, y, by = 1) returns an array from x to y by step = 1
vect <- seq(10, 15)
print(vect)
vect2 <- seq(11, 17, by=2)
print(vect2)
vect3 <- seq(1, 2, by=0.2)
print(vect3)
```

And there are some useful functions for vectors such as `sort()` and `rev()` and `unique()`.

```r
vect <- c(10, 20, 12, 14, 13, 16, 9, 12)
print(sort(vect))
print(rev(vect)) # reverse
print(unique(vect))
```

# Recycling

* Remember when I said that single values are actually vectors of the length 1? What is
really happening here is that R is told to perform element-wise addition on a vector
of length 7 and a vector of length 1. Since element-wise addition is not defined for
vectors of differing lengths, **R recycles the smaller vector until it reaches the same
length as that of the bigger vector**. Once both the vectors are the same size, then R,
element-by-element, performs the addition and returns the result.

```r
print(vect + 3)
```

```r
vect <- c(1, 0, 1, 0, 1, 0)
print(vect + c(FALSE, TRUE))
vect <- c(1, 0, 1, 0, 1)
print(vect + c(FALSE, TRUE))
```

# Anti-Loops

We are going to show why loops are frowned upon by an example.

* If “age” is lower than 250, ageGroup is 1

* If “age” is greater than 250 and lower than 900, ageGroup is 2

* If “age” is greater than 900, age group is 3

```r
df$ageGroup <- 0
print(head(df))
```

## Using for statement

```r
for (i in 1:length(df$age)) {
    if (df$age[i] < 250) 
        df$ageGroup[i] <- 1
    else if (df$age[i] < 900) 
        df$ageGroup[i] <- 2
    else 
        df$ageGroup[i] <- 3
}
print(head(df))
```

## Using Ifelse (Vectorized Function)

```r
# ifelse(Condition, “True Value”, “False Value”)
df <- Orange
df$ageGroup <-
    ifelse(df$age < 250, 1,
        ifelse(df$age < 900, 2, 3)
    )
print(head(df))
```

## Using lapply (Vectorized Function)

```r
# lapply(“Vector/List”,Function)
func <- function(x) {
    if(x < 250)
        return(1)
    else if(x < 900)
        return(2)
    else
        return(3)
}
df$ageGroup <- lapply(df$age, func)
print(head(df))
```
